# Example: Group by a categorical column and compute the mean of a numerical column

```python
grouped_data = df.groupby('Species')['SepalLengthCm'].mean()
print("Grouped data:\n", grouped_data)
```

```python
grouped_data = df.groupby('Species')['SepalWidthCm'].mean()
print("Grouped data:\n", grouped_data)
```

```python
grouped_data = df.groupby('Species')['PetalLengthCm'].mean()
print("Grouped data:\n", grouped_data)
```

```python
grouped_data = df.groupby('Species')['PetalWidthCm'].mean()
print("Grouped data:\n", grouped_data)
```

```python
plt.plot(df['Species'], df['SepalLengthCm'], df['SepalWidthCm'], df['PetalLengthCm'], df['PetalWidthCm'])
plt.title('Iris Species')
plt.xlabel('Species')
plt.ylabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.ylabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')

plt.show()
```

```python
df.groupby('Species')['SepalLengthCm'].mean().plot(kind='bar')
plt.title('Average Sepal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

```python
df.groupby('Species')['SepalWidthCm'].mean().plot(kind='bar')
plt.title('Average Sepal Width per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

```python
df.groupby('Species')['PetalLengthCm'].mean().plot(kind='bar')
plt.title('Average Petal Width per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

```python
df.groupby('Species')['PetalWidthCm'].mean().plot(kind='bar')
plt.title('Average Petal Width per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```
